# Implement a simple Trading Strategy

Simple Moving Average (SMA) Crossover Strategy
Objective: The primary goal of this trading strategy is to identify potential buying (long) and selling (short) opportunities based on the crossovers of two simple moving averages: a short-term SMA and a long-term SMA.

Components of the Strategy:

Short-Term SMA: This is a moving average calculated over a relatively short period, such as 50 days. It responds quickly to price changes and is often used to capture short-term trends.

Long-Term SMA: This is a moving average calculated over a longer period, typically 200 days. It is slower to respond to price changes and is used to identify longer-term trends.

Strategy Rules:

When the short-term SMA crosses above the long-term SMA, it generates a Buy (Long) signal. This suggests that a potential upward trend might be starting, and it's a favorable time to enter a long position.

Conversely, when the short-term SMA crosses below the long-term SMA, it generates a Sell (Short) signal. This indicates a potential downward trend, and it's a suitable time to consider a short position.

When there is no crossover, the strategy advises to Hold, as there is no clear trend change.

How It Works:

Let's say you're trading a stock, and you're using a 50-day SMA and a 200-day SMA. When the 50-day SMA crosses above the 200-day SMA, you receive a "Buy (Long)" signal, indicating a potential bullish trend. This suggests that it might be a good time to enter a long position in that stock.

Conversely, when the 50-day SMA crosses below the 200-day SMA, you receive a "Sell (Short)" signal, indicating a potential bearish trend. This suggests that it might be a suitable time to consider a short position or exit any existing long positions.

Risk Management and Further Analysis:

While the SMA crossover strategy can be a valuable tool, it's important to incorporate risk management techniques and additional analysis. You should always consider factors like stop-loss orders, position sizing, and ovjustments to your approach.

```python
import yfinance as yf
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt

# Access stock data from Yahoo Finance
def get_stock_data(symbol, start_date, end_date):
    stock_data = yf.download(symbol, start=start_date, end=end_date)
    return stock_data

# Implement a simple moving average crossover trading strategy
def simple_moving_average_crossover(stock_data, short_window, long_window):
    signals = pd.DataFrame(index=stock_data.index)
    signals['Signal'] = 0.0

    # Create short simple moving average
    signals['Stock Price'] = stock_data['Close']
    signals['Short_MA'] = stock_data['Close'].rolling(window=short_window, min_periods=1, center=False).mean()

    # Create long simple moving average
    signals['Long_MA'] = stock_data['Close'].rolling(window=long_window, min_periods=1, center=False).mean()

    # Generate signals
    signals['Signal'][short_window:] = np.where(signals['Short_MA'][short_window:] > signals['Long_MA'][short_window:], 1.0, 0.0)

    # Create positions based on signals
    signals['Position'] = signals['Signal'].diff()

    return signals

# Calculate performance metrics
def calculate_performance_metrics(signals, stock_data):
    signals['Trade_Result'] = signals['Position'] * stock_data['Close'].diff()
    signals['Cumulative_Returns'] = signals['Trade_Result'].cumsum()

    # Additional metrics can be calculated here as needed

    return signals

# Visualize the strategy and performance metrics
def visualize_strategy(signals, stock_data):
    plt.figure(figsize=(12, 6))
    plt.plot(stock_data.index, stock_data['Close'], label='Stock Price', alpha=0.7)
    plt.plot(signals.index, signals['Short_MA'], label='Short MA', linestyle='--')
    plt.plot(signals.index, signals['Long_MA'], label='Long MA', linestyle='--')
    plt.title('Simple Moving Average Crossover Strategy')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    # Set your stock symbol, start date, and end date
    symbol = "AMZN"
    start_date = dt.datetime(2022, 1, 1)
    end_date = dt.datetime(2023, 9, 21)

    # Set the short and long moving average windows
    short_window = 50
    long_window = 200

    # Get stock data
    stock_data = get_stock_data(symbol, start_date, end_date)

    # Implement the trading strategy
    trading_signals = simple_moving_average_crossover(stock_data, short_window, long_window)

    # Calculate and visualize performance metrics
    trading_signals = calculate_performance_metrics(trading_signals, stock_data)
    visualize_strategy(trading_signals, stock_data)
```

```python
import gspread
import oauth2client
from gspread_dataframe import set_with_dataframe
import pandas as pd


# Initialize the Google Sheets API
def initialize_google_sheets():
    gc = gspread.service_account(filename="creds.json")
    return gc

# Calculate Profit and Loss (PNL)
def calculate_pnl(row):
    if row['Position'] > 0:
        return round(row['Trade_Result'], 2)
    elif row['Position'] < 0:
        return round(-row['Trade_Result'], 2)
    else:
        return 0.0

# Calculate Gross Profit or Loss
def calculate_gross_pnl(row):
    return round(row['Trade_Result'], 2)

# Calculate Net Profit or Loss
def calculate_net_pnl(row):
    return round(row['Gross PNL'] - row['Expenses'], 2)

# Calculate Return on Investment (ROI)
def calculate_roi(row):
    if row['Position'] != 0:
        return round((row['Net PNL'] / (row['Entry Price'] * abs(row['Position']))) * 100, 2)
    else:
        return 0.0

# Log trading journal in Google Sheets
def log_trading_journal(gc, trading_signals):
    # Open the spreadsheet
    sheet = gc.open('TradingJournal')
    # Get the worksheet
    worksheet = sheet.worksheet('journal1')

    data_to_append = []  # Create an empty list to store the data

    for index, row in trading_signals.iterrows():
        date = index.strftime('%Y-%m-%d')
        action = "Buy (Long)" if row['Position'] > 0 else ("Sell (Short)" if row['Position'] < 0 else "Hold")
        reason = "Signal" if row['Signal'] != 0 else "No Trade"
        entry_price = round(row['Stock Price'], 2)  # Round to 2 decimal places
        exit_price = round(entry_price + row['Trade_Result'], 2)  # Round to 2 decimal places
        position_size = abs(row['Position'])
        gross_pnl = calculate_gross_pnl(row)
        expenses = 0.0  # You can add expenses calculation here
        net_pnl = calculate_net_pnl({'Gross PNL': gross_pnl, 'Expenses': expenses})
        roi = calculate_roi({'Net PNL': net_pnl, 'Entry Price': entry_price, 'Position': row['Position']})

        data_to_append.append([date, action, reason, entry_price, exit_price, position_size, gross_pnl, expenses, net_pnl, roi])

    # Convert the data to a Pandas DataFrame
    df = pd.DataFrame(data_to_append, columns=['Date', 'Action', 'Reason', 'Entry Price', 'Exit Price', 'Position Size', 'Gross PNL', 'Expenses', 'Net PNL', 'ROI'])

    # Write the DataFrame to the worksheet

    
    # Write the DataFrame to the worksheet starting from cell A1
    set_with_dataframe(worksheet, df)
    
    # Save the spreadsheet
    #worksheet.save()

if __name__ == "__main__":
    # Initialize Google Sheets API
    gc = initialize_google_sheets()

    # Set your Google Sheets document name
    sheet_name = "TradingJournal"  # Replace with your Google Sheets document name

    # Log trading journal in Google Sheets
    log_trading_journal(gc, trading_signals)
    print("Trading journal logged in Google Sheets.")
```
